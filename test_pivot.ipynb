```python
import os

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def split_dataframe(df, chunk_size = 10000): 
    chunks = list()
    if len(df) % chunk_size == 1:
        num_chunks = len(df) // chunk_size + 1
    else:
        num_chunks = len(df) // chunk_size
        
    for i in range(num_chunks):
        chunks.append(df[i*chunk_size:(i+1)*chunk_size])
    return chunks
```

```python
df3 = pd.DataFrame({'customer_id': ['c1', 'c2', 'c2', 'c3', 'c4', 'c1'],
                    'product_id': ['p1', 'p2', 'p3', 'p4', 'p5', 'p5'],
                    'review_score': [1, 2, 3, 4, 5, 2],
                    })

df3_pivot = df3.pivot_table(index='customer_id', columns='product_id', values='review_score', aggfunc=np.sum, fill_value=None)
df3_chunks = split_dataframe(df3, chunk_size=2)

total = pd.DataFrame(columns=df3['product_id'].unique(), index=df3['customer_id'].unique())

# df3_chunk_pivots = []
for chunk in df3_chunks:
    pivot_chunk = chunk.pivot(index='customer_id', columns='product_id', values='review_score')
    total.add(pivot_chunk)
    display(pivot_chunk)

display(total)
```
