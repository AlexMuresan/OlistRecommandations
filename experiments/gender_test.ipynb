```python
import cv2
import names
import urllib
import requests
import numpy as np
import matplotlib.pyplot as plt

from skimage import io
from skimage.transform import rescale
```

```python
def generate_people(min_id, max_id):
    random_people = {}
    gender_dict = {1: "male", 2: "female"}

    for i in range(min_id, max_id + 1):
        gender = gender_dict[np.random.randint(1, 3)]
        random_name = names.get_full_name(gender=gender)
        age = np.random.randint(18, 75)

        # while random_name in random_people.values():
        #     random_name = names.get_full_name(gender=gender)
        url_string = f'https://fakeface.rest/face/json?minimum_age={age}&maximum_age={age}&gender={gender}'
        response = requests.get(url_string)
        response.encoding = 'utf-8'
        url = response.json()['image_url']

        random_people[i] = {
            "name": random_name,
            "image": rescale(io.imread(url), 0.25, channel_axis=-1)}

    return random_people
```

```python
people = generate_people(1, 10)
```
