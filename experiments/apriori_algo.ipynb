```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
transactions = pd.read_csv("./store_data.csv", header=None)
transactions.head()
```

```python
no_transactions = len(transactions)
max_items_per_transactions = len(transactions.columns)
records = []
for i in range(0, no_transactions):
    records.append([
        str(transactions.values[i, j]) 
        for j in range(max_items_per_transactions)])

print(records[0])
```

```python
association_rules = apriori(
    records, 
    min_support=0.0045, 
    min_confidence=0.2,
    min_lift=3,
    min_length=2)

association_results = list(association_rules)
print(f"Identified {len(association_results)} association rules")
```

```python
for idx, item in enumerate(association_results):
    pair = item[0]
    items = [x for x in pair]
    print(f"Rule: {items[0]} -> {items[1]}")
    print(f"Support: {item[1]}")
    print(f"Confidence: {item[2][0][2]:0.3f}")
    print(f"Lift: {item[2][0][3]:0.3f}")
    print("===============================")

    if idx == 4:
        break
```
