```python
import random
import numpy as np
import pandas as pd

from IPython.display import display, Math
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def print_matrix(array):
    matrix = ''
    for row in array:
        try:
            for number in row:
                matrix += f'{number:.2f}&'
        except TypeError:
            matrix += f'{row}&'
        matrix = matrix[:-1] + r'\\'
    display(Math(r'\begin{bmatrix}'+matrix+r'\end{bmatrix}'))
```

```python
pd.options.display.float_format = '{:.2f}'.format


nan_matrix = np.full((5, 6), np.NaN)
rows, cols = nan_matrix.shape
for i in range(rows):
    col_places = [random.randint(0, cols-1) for _ in range(random.randint(3, int(cols/2)))]
    for col in col_places:
        nan_matrix[i][col] = random.randint(1, 5)

nan_df = pd.DataFrame(nan_matrix)
display(nan_df)

nan_filled = nan_df.fillna(nan_df.mean(axis=0))
display(nan_filled)

nan_cosine = cosine_similarity(nan_filled.values)
print_matrix(nan_cosine)
```
